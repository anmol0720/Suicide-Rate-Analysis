# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/home/nbuser/library/WEEK3/master.csv')
```

## Short Glimps On What We Are Going To Work On

```python
df.head(7)
```

# Substituting Not a Number Value

```python
df['HDI for year']=df['HDI for year'].mean()
df.head(7)
```

# A Relation between each and every Variable
    Heat Map Correlation Display

```python
corr=df.corr()
sns.heatmap(corr)
```

    The Darker The shade the more correlated it is.

# Dependence of Suicide/100k pop according to different Country

```python
df1=df.groupby(['country'])['country','suicides/100k pop'].mean()
df1['country']=df1.index
df1.plot.bar(figsize=(20,10))
plt.title('Country vs suicides/100k pop')
plt.ylabel('Avg. suicides/100k pop value')
plt.xlabel('Country')
```

    With The help of above Graph We Can Distinguish Between Different Countries On the basis of suicides/100k pop

# Dependence of Suicide/100k pop according to different Years
    The Below Displayed is the Scatter Plot For Clear Visualisation Of the relation between two variables

```python
df2=df.groupby(['year'])['year','suicides/100k pop'].mean()
df2['year']=df2.index
plt.scatter(df2['year'],df2['suicides/100k pop'])
plt.title('Year vs suicides/100k pop')
plt.ylabel('Avg. suicides/100k pop value')
plt.xlabel('Year')
```

# Relation between Suicide/100k pop and Sex

```python
df3=df.groupby(['sex'])['sex','suicides/100k pop'].mean()
df3.plot.bar()
plt.title('Sex vs suicides/100k pop')
plt.ylabel('Avg. suicides/100k pop')
plt.xlabel('Sex')
```

    We can conclude from the above bar plot that the Suicides/100k pop is more for male as compared to female

# Dependence of Suicide/100k pop on Suicides Number
    The Below Display Relation Between Suicides/10k pop and Suicides Numbers In Relation To Different Sex

```python
sns.lmplot('suicides_no','suicides/100k pop',data=df,hue='sex',fit_reg=False)
plt.title('suicides number vs suicides/100k pop')
```

# Dependence of Suicide/100k pop on Population
    The Below Display Relation Between Suicides/10k pop and Population In Relation To Different Sex

```python
sns.lmplot('population','suicides/100k pop',hue='sex',data=df,fit_reg=False)
```

# Relation between Suicide/100k and differnet Age Groups

```python
df5=df.groupby(['age'])['age','suicides/100k pop'].mean()
df5.plot()
plt.title('Age Group vs Suicides/100k pop')
plt.xlabel('Age')
plt.ylabel('Avg. suicides/100k pop')
```

# Dependence of Suicide/100k pop on differnet Generations

```python
df6=df.groupby(['generation'])['generation','suicides/100k pop'].mean()
df6.plot.bar()
plt.xlabel('Generation')
plt.ylabel('Avg. suicides/100k pop')
```

    From The Above Bar Graph We Can Conclude That G.I. Generation Has Moost Suicides/100k pop
